```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import time

# Define the date range for data collection
data_from = '03/16/2024'
today_date = datetime.now()
data_to = today_date.strftime('%m/%d/%Y')

url = 'http://oasis.caiso.com/mrioasis/default.do?tiny=izbRTw'
cb = webdriver.Chrome()

cb.get(url)
time.sleep(5)  # Wait for the page to load

# Use WebDriverWait to wait for the page elements to load
wait = WebDriverWait(cb, 30)  # Adjust timeout as needed

# Navigate to "SYSTEM DEMAND"
sys_page = wait.until(EC.element_to_be_clickable((By.XPATH, './/*[@data-copy-text="SYSTEM DEMAND"]')))
ActionChains(cb).move_to_element(sys_page).perform()

# Select "Wind and Solar Forecast"
wind_page = wait.until(EC.element_to_be_clickable((By.XPATH, './/*[@data-copy-text="Wind and Solar Forecast"]')))
wind_page.click()

# Input the "data_to" date
time.sleep(0.5)
data_to_input = wait.until(EC.element_to_be_clickable((By.XPATH, '//input[contains(@onchange, "PFC_date_to")]')))
data_to_input.clear()  # Clear the field before sending keys
data_to_input.send_keys(data_to)

# Input the "data_from" date
time.sleep(0.5)
data_from_input = wait.until(EC.element_to_be_clickable((By.XPATH, '//input[contains(@onchange, "PFC_date_from")]')))
data_from_input.clear()  # Clear the field before sending keys
data_from_input.send_keys(data_from)

# Click the "Apply" button to submit the date range and retrieve data
time.sleep(0.5)
apply_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[@title="Apply"]')))
apply_button.click()


time.sleep(10)
# Wait for the "Download CSV" button to become clickable and click it to download the data
download_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[@title="Download CSV"]')))
download_button.click()

# Optionally, wait for the download to complete. This approach depends on your browser's download behavior.

time.sleep(60)
cb.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from datetime import datetime
import time

# Define the date range for data collection
data_from = '03/16/2024'
today_date = datetime.now()
data_to = today_date.strftime('%m/%d/%Y')

url = 'http://oasis.caiso.com/mrioasis/default.do?tiny=izbRTw'
cb = webdriver.Chrome()

cb.get(url)
time.sleep(5)  # Wait for the page to load

# Navigate to "SYSTEM DEMAND" 
sys_page = cb.find_element(by=By.XPATH, value='.//*[@data-copy-text="SYSTEM DEMAND"]')
ActionChains(cb).move_to_element(sys_page).perform()
time.sleep(0.5)

# Select "Wind and Solar Forecast" 
wind_page = cb.find_element(by=By.XPATH, value='.//*[@data-copy-text="Wind and Solar Forecast"]')
ActionChains(cb).move_to_element(wind_page).perform()
wind_page.click()
time.sleep(5)


data_to_input = cb.find_element(by=By.XPATH, value='//input[contains(@onchange, "PFC_date_to")]')
ActionChains(cb).move_to_element(data_to_input).perform()
data_to_input.send_keys(Keys.END + Keys.BACK_SPACE * 10)
time.sleep(0.5)
data_to_input.send_keys(data_to)
time.sleep(0.5)

data_from_input = cb.find_element(by=By.XPATH, value='//input[contains(@onchange, "PFC_date_from")]')
ActionChains(cb).move_to_element(data_from_input).perform()
data_from_input.send_keys(Keys.END + Keys.BACK_SPACE * 10)
time.sleep(0.5)
data_from_input.send_keys(data_from)
time.sleep(0.5)


# Click the "Apply" button
apply_button =  cb.find_element(by=By.XPATH, value='//button[@title="Apply"]')
ActionChains(cb).move_to_element(apply_button).perform()
apply_button.click()
time.sleep(15)

# Click the "Download CSV" button to download the data
download_button =  cb.find_element(by=By.XPATH, value='//button[@title="Download CSV"]')
ActionChains(cb).move_to_element(download_button).perform()
download_button.click()
time.sleep(60)
cb.quit()
```
